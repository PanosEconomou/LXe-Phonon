# Simulation of background phonon radiation in a low temperature system

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as c

plt.style.use('dark_background')

T = np.linspace(70,200,4) #K
N = int(1e6)
w = np.linspace(1,1e4,N)

fig = plt.figure(figsize=(8,6),dpi = 300)
ax = fig.add_subplot(111)

for t in T:
    n = 1/(np.exp(c.hbar*w/(c.k*t))-1)
    ax.plot(w,n,lw=2,label=r'$T = %d\ K$'%int(t))

ax.set_xscale('log')
ax.set_xlabel(r'Phonon Frequency $\omega$ $[rad\ s^{-1}]$')
ax.set_ylabel(r"Mean number of particles $\langle n \rangle$")
ax.set_xlim(0.4e2,w.max())
ax.set_ylim(0,6e11)
ax.legend(frameon=False,fontsize = 20)
```
